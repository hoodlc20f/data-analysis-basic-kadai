```python
# !pip install japanize-matplotlib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import japanize_matplotlib

df = pd.read_csv('sample_pandas_6.csv')
category_df = pd.read_csv('category.csv')
mdf = pd.merge(df, category_df[['商品番号', 'カテゴリー']], how='inner', on='商品番号')

category_cnt = mdf['カテゴリー'].value_counts()
category_cnt.plot(kind='bar')
plt.title('各カテゴリー出現頻度')
plt.show()

df.groupby('商品番号')['注文数'].describe()
```
