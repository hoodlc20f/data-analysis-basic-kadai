```python
import numpy as np
import matplotlib.pyplot as plt

# tight_layout=Trueは、サブプロットのレイアウトをある程度自動で調整するオプション
fig, axes = plt.subplots(2, 3, tight_layout=True)

y0 = [1, 2, -5, 2]
axes[0,0].plot(y0)

x1 = np.linspace(0, 10, 100)
y1 =  np.sin(x1)
axes[0,1].plot(x1,y1)
axes[0,1].set(xlim=(0,10), xticks=np.arange(0,11),
              ylim=(-1,1), yticks=np.arange(-1,2))

x2 = np.random.normal(50, 10, 1000)
axes[0,2].set_xlim(15,80)
axes[0,2].hist(x2, bins=60)

np.random.seed(1)
x3 = np.random.normal(5, 5, 1000)
y3 = np.random.normal(5, 5, 1000) * x3
axes[1,0].scatter(x3,y3)

plt.show()
```
